# 0. Import các thư viện

```python
import pandas as pd
```

# 1. Đọc dữ liệu từ file CSV

```python
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31020936/GDPlist.csv", encoding = "ISO-8859-1")
df.head()
```

```python
df.info()
```

# 2. Việt hoá tên các cột trong bảng dữ liệu

```python
# Đổi tên cột "Country" thành "Nuoc", "Continent" thành "Chau luc" và "GDP (millions of US$)" thành "GDP (trieu $)"

df.rename(columns = {"Country": "Nuoc",
                         "Continent": "Chau luc",
                         "GDP (millions of US$)": "GDP (trieu $)"}, inplace = True)
df.head()
```

# 3. Chèn thêm một cột "Thanh pho" sau cột "Nuoc", giá trị ban đầu là giá trị của cột "Nuoc"

```python
df.insert(1,"Thanh pho", pd.Series(df.Nuoc, index = df.index))
df.head()
```

# 5. Xoá các bản ghi có châu lục là "Asia"

```python
# Xác định chỉ số hàng
row_index = df.loc[df["Chau luc"] == "Asia"].index
row_index
```

```python
df.drop(row_index, inplace = True)
df
```

# 6. Xoá các bản ghi có GDP < 300000

```python
# Xác định chỉ số hàng
row_index = df.loc[df["GDP (trieu $)"] < 300000].index
row_index
```

```python
# Sử dụng phương thức drop() để xoá các hàng chứa GDP < 300000
df.drop(row_index, inplace = True)
df
```
